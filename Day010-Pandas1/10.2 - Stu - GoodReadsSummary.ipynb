```python
# File to Load

csvpath = "../Resources/NewBooks.csv"
# Read the modified GoodReads csv and store into Pandas DataFrame
df = pd.read_csv(csvpath)
df.head()
```

```python
# Calculate the number of unique authors in the DataFrame
numberOfAuthors = len(df["Authors"].unique())
print(numberOfAuthors)
# Calculate the earliest/latest year a book was published
earliest = df["Year"].min()
latest = df["Year"].max()
print(earliest,latest)
# Calculate the total reviews for the entire datase
total = df["One Star"].sum() + df["Two Star"].sum() 
total
```

```python
# Place all of the data found into a summary DataFrame
summary = pd.DataFrame(
{
    "Number of Authors": [numberOfAuthors],
    "Earliest Pub": [earliest],
    "Latest Pub": [latest]
})
summary
```
