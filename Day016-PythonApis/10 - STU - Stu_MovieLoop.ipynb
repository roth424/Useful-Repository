```python
# Dependencies
import requests
from pprint import pprint

url = "http://www.omdbapi.com/?apikey=trilogy&t="

movies = ["Some Kind of Wonderful", "Rollerball", "Cloudy with a Chance of Meatballs"]

movieList = []
directors = []

for movie in movies:
    data = requests.get(url + movie).json()
    # print(data)
    movieList.append(data)
    directors.append(data["Director"])
    
pprint(movieList)
print("\n\n")
for i in range(len(directors)):
    print(f"Movie: {movies[i]} was directed by {directors[i]}")
```
