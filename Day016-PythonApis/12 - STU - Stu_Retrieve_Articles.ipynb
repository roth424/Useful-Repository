```python
# Dependencies
from pprint import pprint
import requests
from config import nyt_api_key
import time

baseurl = "https://api.nytimes.com/svc/search/v2/articlesearch.json?api-key=" + nyt_api_key + "&"

# Store a search term
searchFor = "debate"

# Search for articles published between a begin and end date
beginDate = '20190601'
endDate = '20190629'

# Build url
url = baseurl + "q=" + searchFor + "&begin_date=" + beginDate + "&end_date=" + endDate
print(url)
```

```python
# Retrieve articles
articles = []
for i in range(5):
    data = requests.get(url + "&page=" + str(i)).json()
    for article in data["response"]["docs"]:
        articles.append(article)
    print(len(articles))
    time.sleep(5)
pprint(articles)
```

```python
for article in articles:
    print(article["headline"]["main"])
```
